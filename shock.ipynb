```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def sigmoid(x):
    return 1 / (1 + np.exp(-5 * x))

# Generate x values
x_values = np.linspace(-3, 3, 100)  # 100 points from -3 to 3
y_values = sigmoid(x_values)          # Calculate y values using the sigmoid function

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label=r'$y = \frac{1}{1 + e^{-5x}}$', color='blue')
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black',linewidth=0.5, ls='--')  # X-axis
plt.axvline(0, color='black',linewidth=0.5, ls='--')  # Y-axis
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.ylim(-0.1, 1.1)                       # Set y-axis limits for better visibility
plt.xlim(-3, 3)                          # Set x-axis limits
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return 3 * np.exp(-x) * np.sin(2 * x)

# Generate x values
x_values = np.linspace(0, 10, 1000)  # 1000 points from 0 to 10
y_values = f(x_values)                 # Calculate y values using the function

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label=r'$y = 3e^{-x} \sin(2x)$', color='blue')
plt.title('Plot of $y = 3e^{-x} \sin(2x)$')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=0.5, ls='--')  # X-axis
plt.axvline(0, color='black', linewidth=0.5, ls='--')  # Y-axis
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.ylim(-4, 4)  # Set y-axis limits for better visibility
plt.xlim(0, 10)  # Set x-axis limits
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of x values for both functions
x_demand = np.linspace(-10, 10, 500)  # Range for demand function
x_combined = np.linspace(-10, 10, 500)  # Range for combined function

# Define the demand function
D = np.exp(0.1 * x_demand)  # e^(0.1x)

# Define the combined function
def additive_combination(x):
    return part_one(x) + part_two(x)

def part_one(x):
    return 3 * np.exp(-x) * np.sin(2 * x)

def part_two(x):
    return (1 / (1 + np.exp(-5 * x)))

# Generate y values for the combined function
y_combined = additive_combination(x_combined)

D_combined = np.exp(0.1 * x_combined) * y_combined


part_one_sim = part_one(x_combined)
part_two_sim = part_two(x_combined)


# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x_demand, D, label='Exponential demand', color='blue')
plt.plot(x_combined, y_combined, label='External shock simulation', color='orange')
plt.plot(x_combined, part_one_sim, label='First term', color='green', linestyle='--')
plt.plot(x_combined, part_two_sim, label='Second Term', color='green', linestyle='--')
plt.title('Demand Function and Combined Function')
plt.xlabel('Time (x) in years')
plt.ylabel('Value')
plt.legend()
plt.grid()
plt.xlim(-10, 10)  # Set x-axis limits from -10 to 10
plt.ylim(-3, np.max(D) * 1.1)  # Adjust y-axis limit for better visibility
plt.show()
```
